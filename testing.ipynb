```python
import cv2, os
import matplotlib.pyplot as plt
```

```python
root_dir = "./PolypGen"
train_images = os.listdir(os.path.join(root_dir, "train/images"))
train_masks = os.listdir(os.path.join(root_dir, "train/masks"))
train_images = [os.path.join(root_dir, "train/images", img) for img in train_images]
train_masks = [os.path.join(root_dir, "train/masks", mask) for mask in train_masks]
```

```python
plt.figure(figsize=(15, 8))
for idx, img_path in enumerate(train_images[:5]):
    image = cv2.imread(img_path)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    plt.subplot(1, 5, idx + 1)
    plt.imshow(image)
    plt.title(os.path.basename(img_path))
    plt.axis('off')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15, 8))
for idx, img_path in enumerate(train_images[:5]):
    image = cv2.imread(img_path)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    image = cv2.resize(image, (512, 512), interpolation=cv2.INTER_LANCZOS4)
    plt.subplot(1, 5, idx + 1)
    plt.imshow(image)
    plt.title(os.path.basename(img_path))
    plt.axis('off')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15, 8))
for idx, img_path in enumerate(train_masks[:5]):
    train_mask_orig = cv2.imread(img_path)
    train_mask_orig = cv2.cvtColor(train_mask_orig, cv2.COLOR_BGR2RGB)
    plt.subplot(1, 5, idx + 1)
    plt.imshow(train_mask_orig)
    plt.title(os.path.basename(img_path))
    plt.axis('off')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15, 8))
for idx, img_path in enumerate(train_masks[:5]):
    train_mask_tr = cv2.imread(img_path)
    train_mask_tr = cv2.cvtColor(train_mask_tr, cv2.COLOR_BGR2RGB)
    train_mask_tr = cv2.resize(train_mask_tr, (512, 512), interpolation=cv2.INTER_LANCZOS4)
    plt.subplot(1, 5, idx + 1)
    plt.imshow(train_mask_tr)
    plt.title(os.path.basename(img_path))
    plt.axis('off')
plt.tight_layout()
plt.show()
```
